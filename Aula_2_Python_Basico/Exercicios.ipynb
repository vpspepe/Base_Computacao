## Resoluções

```python
# 1. Mostre a mensagem "Bem vindo apostador!" na tela.
print("Bem vindo apostador!")

# 2. Peça um valor apostado e então mostre a mensagem O valor apostado foi [valor].
valor_apostado = float(input("Digite o valor apostado: "))
print(f"O valor apostado foi {valor_apostado}")

# 3. Peça dois valores apostados e imprima a soma.
valor1 = float(input("Digite o primeiro valor apostado: "))
valor2 = float(input("Digite o segundo valor apostado: "))
print(f"A soma dos valores apostados é {valor1 + valor2}")

# 4. Peça os valores apostados em 4 semanas e mostre a média.
valores = [float(input(f"Digite o valor apostado na semana {i+1}: ")) for i in range(4)]
media = sum(valores) / len(valores)
print(f"A média dos valores apostados é {media}")

# 5. Converta o valor apostado em dólares para reais.
cotacao_dolar = float(input("Digite a cotação do dólar: "))
valor_em_dolares = float(input("Digite o valor apostado em dólares: "))
valor_em_reais = valor_em_dolares * cotacao_dolar
print(f"O valor apostado em reais é {valor_em_reais}")

# 6. Peça o valor apostado em uma partida, calcule e mostre o possível retorno com base em uma odd.
valor_apostado = float(input("Digite o valor apostado: "))
odd = float(input("Digite a odd: "))
retorno = valor_apostado * odd
print(f"O possível retorno é {retorno}")

# 7. Calcule o valor total apostado em um mês, em seguida mostre o dobro deste valor para o usuário.
valores_mes = [float(input(f"Digite o valor apostado na semana {i+1}: ")) for i in range(4)]
total_mes = sum(valores_mes)
print(f"O dobro do valor total apostado no mês é {total_mes * 2}")

# 8. Pergunte quanto você ganhou por aposta e o número de apostas feitas no mês. Calcule e mostre o total do seu ganho no referido mês.
ganho_por_aposta = float(input("Digite quanto você ganhou por aposta: "))
numero_apostas = int(input("Digite o número de apostas feitas no mês: "))
total_ganho = ganho_por_aposta * numero_apostas
print(f"O total do seu ganho no mês é {total_ganho}")

# 9. Peça o valor ganho em uma aposta em dólares, transforme e mostre o valor ganho em reais.
cotacao_dolar = float(input("Digite a cotação do dólar: "))
valor_ganho_dolares = float(input("Digite o valor ganho em dólares: "))
valor_ganho_reais = valor_ganho_dolares * cotacao_dolar
print(f"O valor ganho em reais é {valor_ganho_reais}")

# 10. Peça o valor ganho em uma aposta em reais, transforme e mostre em dólares.
cotacao_dolar = float(input("Digite a cotação do dólar: "))
valor_ganho_reais = float(input("Digite o valor ganho em reais: "))
valor_ganho_dolares = valor_ganho_reais / cotacao_dolar
print(f"O valor ganho em dólares é {valor_ganho_dolares}")

# 11. Peça 2 valores apostados inteiros e um valor ganho real. Calcule e mostre:
valor1 = int(input("Digite o primeiro valor apostado: "))
valor2 = int(input("Digite o segundo valor apostado: "))
valor_ganho = float(input("Digite o valor ganho: "))

# o produto do dobro do primeiro valor apostado com metade do segundo.
produto = (2 * valor1) * (valor2 / 2)
print(f"O produto do dobro do primeiro valor apostado com metade do segundo é {produto}")

# a soma do triplo do primeiro valor apostado com o valor ganho.
soma = (3 * valor1) + valor_ganho
print(f"A soma do triplo do primeiro valor apostado com o valor ganho é {soma}")

# o valor ganho elevado ao cubo.
valor_ganho_cubo = valor_ganho ** 3
print(f"O valor ganho elevado ao cubo é {valor_ganho_cubo}")

# 12. Calcule o possível retorno com base na fórmula: valor_apostado * odd
valor_apostado = float(input("Digite o valor apostado: "))
odd = float(input("Digite a odd: "))
retorno = valor_apostado * odd
print(f"O possível retorno é {retorno}")

```

# **Desafio**

- Faça um Programa que pergunte quanto você ganhou em apostas esportivas online no mês e o valor total apostado. Calcule e mostre o lucro líquido no referido mês, sabendo-se que são descontados 10% para a plataforma de apostas, 5% para taxas de transação e 2% para impostos, faça um programa que nos dê:

    - - quanto pagou para a plataforma de apostas.
    - - quanto pagou em taxas de transação.
    - - quanto pagou em impostos.
    - - o lucro líquido.
    - - lucro bruto.
    - - calcule os descontos e o lucro líquido, conforme a tabela abaixo:

    + Lucro Bruto : R$
    - Plataforma de Apostas (10%) : R$
    - Taxas de Transação (5%) : R$
    - Impostos (2%) : R$
    = Lucro Líquido : R$

    Obs.: Lucro Bruto - Descontos = Lucro Líquido.

Faça isso para N meses

## Resolução

```python
n_meses: int = int(input("Digite o número de meses: "))
mes_atual: int = 0
lucro_bruto_total: float = 0
apostado_total: float = 0

while (n_meses > mes_atual):
    lucro_bruto: float = float(input(f"Quanto voce ganhou no mes {mes_atual}? "))
    apostado_mes: float = float(input(f"Quanto você apostou no mês {mes_atual}?"))
    if lucro_bruto > 0:
        imposto_plataforma: float = 0.1 * lucro_bruto
        taxa_transasacao: float = 0.05 * lucro_bruto
        imposto: float = 0.02 * lucro_bruto
        lucro_liquido: float = lucro_bruto - imposto_plataforma - taxa_transasacao - imposto
        print(f"Voce pagou R$ {imposto_plataforma} para a platafoma!")
        print(f"Voce pagou R$ {taxa_transasacao} em taxas transacionais!")
        print(f"Voce pagou R$ {imposto} para o governo!")
        print(f"Seu lucro liquido foi de R$ {lucro_liquido}")
        print(f"Seu lucro bruto foi R$ {lucro_bruto}")
    else: 
        print("Você não teve lucro esse mês!")
        prejuizo: float = -1 * lucro_bruto
        print(f"Seu prejuizo foi de R$ {prejuizo}")
    lucro_bruto_total += lucro_bruto
    apostado_total += apostado_mes
    mes_atual += 1
    
print(f"Seu lucro bruto total foi de R$ {lucro_bruto_total}")
print(f"Você apostou um total de R$ {apostado_total}")
percentual:float = lucro_bruto_total / apostado_total
if lucro_bruto_total > 0:
    print(f"Seu lucro percentual total foi de", percentual)
else:
    print("Seu prejuízo percentual total foi de", -1*percentual)
```

Verificação de Login:

O usuário inicialmente cadastra 3 contas indicando:
- Nome
- Senha
- Login
- Tipo (Básico, Prata ou Premium)
- Saldo inicial

Após cadastrar as 3 contas, será solicitado Login e Senha para entrar em alguma das 3 contas. Se nenhuma conta corresponder, deve ser impressa uma mensagem de erro e o usuário deve poder tentar novamente. Nessa mensagem de erro, se o usuário apenas apertar "Enter", ele tenta de novo, mas caso ele digite "Sair" e depois pressionar "Enter", o programa encerra.

Caso ele faça Login, serão solicitadas as Tips do usuário, indicando:
- Valor Apostado
- Valor Ganho

Ao final de cada Tip, aparece uma mensagem perguntando se ele quer continuar. Se ele não quiser informar mais nenhuma Tip, ele deve digitar "Sair", senão só aperta Enter.
Após informar todas as Tips, o programa calcula o lucro líquido do usuário (Lembre o que acontece quando há prejuízo!).
Cálculo de Lucro:
- Lucro Bruto = Valor Ganho - Valor Apostado
- Lucro Líquido = Lucro Bruto - Imposto * (1 - Fator de redução)
- Imposto:
- - 5% do lucro bruto se este for menor que R$ 100,00
- - 15% do lucro bruto se este for menor que R$ 1000,00 e maior que R$ 100,00
- - 20% do lucro bruto se este for maior que R$ 1000,00
- Fator de redução: 
- - 0% para Básico
- - 5% para Prata 
- - 10% para Premium

Ao final, deve ser impresso:
- Nome do usuário
- Lucro Líquido Total
- Valor Total Apostado
- % de Tips bem sucedidas
- Imposto Pago

```python
# Cadastro de contas
nome1 = input("Digite o nome do usuário 1: ")
senha1 = input("Digite a senha do usuário 1: ")
login1 = input("Digite o login do usuário 1: ")
tipo1 = input("Digite o tipo do usuário 1 (Básico, Prata ou Premium): ")
saldo1 = float(input("Digite o saldo inicial do usuário 1: "))

nome2 = input("Digite o nome do usuário 2: ")
senha2 = input("Digite a senha do usuário 2: ")
login2 = input("Digite o login do usuário 2: ")
tipo2 = input("Digite o tipo do usuário 2 (Básico, Prata ou Premium): ")
saldo2 = float(input("Digite o saldo inicial do usuário 2: "))

nome3 = input("Digite o nome do usuário 3: ")
senha3 = input("Digite a senha do usuário 3: ")
login3 = input("Digite o login do usuário 3: ")
tipo3 = input("Digite o tipo do usuário 3 (Básico, Prata ou Premium): ")
saldo3 = float(input("Digite o saldo inicial do usuário 3: "))

# Verificação de login
while True:
    login = input("Digite seu login: ")
    senha = input("Digite sua senha: ")

    if (login == login1 and senha == senha1):
        nome = nome1
        tipo = tipo1
        break
    elif (login == login2 and senha == senha2):
        nome = nome2
        tipo = tipo2
        break
    elif (login == login3 and senha == senha3):
        nome = nome3
        tipo = tipo3
        break
    else:
        print("Login ou senha incorretos. Tente novamente.")
        if input("Digite 'Sair' para encerrar ou pressione Enter para tentar novamente: ") == "Sair":
            exit()

# Solicitação de tips
total_apostado = 0
total_ganho = 0
total_tips = 0
tips_sucesso = 0
total_lucro_liquido = 0
total_imposto = 0

while True:
    valor_apostado = float(input("Digite o valor apostado: "))
    valor_ganho = float(input("Digite o valor ganho: "))

    total_apostado += valor_apostado
    total_ganho += valor_ganho
    total_tips += 1

    sucesso: bool = valor_ganho > valor_apostado
    if sucesso:
        tips_sucesso += 1

    # Cálculo do lucro para cada aposta
    lucro_bruto = valor_ganho - valor_apostado

    if lucro_bruto < 0:
        imposto_base = 0
    elif lucro_bruto <= 100:
        imposto_base = 0.05 * lucro_bruto
    elif lucro_bruto <= 1000:
        imposto_base = 0.15 * lucro_bruto
    else:
        imposto_base = 0.20 * lucro_bruto

    if tipo == "Básico":
        fator_reducao = 0
    elif tipo == "Prata":
        fator_reducao = 0.05
    elif tipo == "Premium":
        fator_reducao = 0.10

    imposto = imposto_base * (1 - fator_reducao)
    lucro_liquido = lucro_bruto - imposto

    total_lucro_liquido += lucro_liquido
    total_imposto += imposto

    print(f"Lucro Líquido da aposta: R$ {lucro_liquido:.2f}")
    print(f"Imposto Pago na aposta: R$ {imposto:.2f}")

    if input("Digite 'Sair' para encerrar ou pressione Enter para continuar: ") == "Sair":
        break

# Impressão dos resultados finais
print('\n', 100*'-', '\n')
print(f"Nome do usuário: {nome}")
print(f"Lucro Líquido Total: R$ {total_lucro_liquido:.2f}")
print(f"Valor Total Apostado: R$ {total_apostado:.2f}")
print(f"% de Tips bem sucedidas: {tips_sucesso / total_tips * 100:.2f}%")
print(f"Imposto Total Pago: R$ {total_imposto:.2f}")
```
