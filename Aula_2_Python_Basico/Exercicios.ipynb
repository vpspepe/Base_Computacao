## Resoluções

```python
# 1. Mostre a mensagem "Bem vindo apostador!" na tela.
print("Bem vindo apostador!")

# 2. Peça um valor apostado e então mostre a mensagem O valor apostado foi [valor].
valor_apostado = float(input("Digite o valor apostado: "))
print(f"O valor apostado foi {valor_apostado}")

# 3. Peça dois valores apostados e imprima a soma.
valor1 = float(input("Digite o primeiro valor apostado: "))
valor2 = float(input("Digite o segundo valor apostado: "))
print(f"A soma dos valores apostados é {valor1 + valor2}")

# 4. Peça os valores apostados em 4 semanas e mostre a média.
valores = [float(input(f"Digite o valor apostado na semana {i+1}: ")) for i in range(4)]
media = sum(valores) / len(valores)
print(f"A média dos valores apostados é {media}")

# 5. Converta o valor apostado em dólares para reais.
cotacao_dolar = float(input("Digite a cotação do dólar: "))
valor_em_dolares = float(input("Digite o valor apostado em dólares: "))
valor_em_reais = valor_em_dolares * cotacao_dolar
print(f"O valor apostado em reais é {valor_em_reais}")

# 6. Peça o valor apostado em uma partida, calcule e mostre o possível retorno com base em uma odd.
valor_apostado = float(input("Digite o valor apostado: "))
odd = float(input("Digite a odd: "))
retorno = valor_apostado * odd
print(f"O possível retorno é {retorno}")

# 7. Calcule o valor total apostado em um mês, em seguida mostre o dobro deste valor para o usuário.
valores_mes = [float(input(f"Digite o valor apostado na semana {i+1}: ")) for i in range(4)]
total_mes = sum(valores_mes)
print(f"O dobro do valor total apostado no mês é {total_mes * 2}")

# 8. Pergunte quanto você ganhou por aposta e o número de apostas feitas no mês. Calcule e mostre o total do seu ganho no referido mês.
ganho_por_aposta = float(input("Digite quanto você ganhou por aposta: "))
numero_apostas = int(input("Digite o número de apostas feitas no mês: "))
total_ganho = ganho_por_aposta * numero_apostas
print(f"O total do seu ganho no mês é {total_ganho}")

# 9. Peça o valor ganho em uma aposta em dólares, transforme e mostre o valor ganho em reais.
cotacao_dolar = float(input("Digite a cotação do dólar: "))
valor_ganho_dolares = float(input("Digite o valor ganho em dólares: "))
valor_ganho_reais = valor_ganho_dolares * cotacao_dolar
print(f"O valor ganho em reais é {valor_ganho_reais}")

# 10. Peça o valor ganho em uma aposta em reais, transforme e mostre em dólares.
cotacao_dolar = float(input("Digite a cotação do dólar: "))
valor_ganho_reais = float(input("Digite o valor ganho em reais: "))
valor_ganho_dolares = valor_ganho_reais / cotacao_dolar
print(f"O valor ganho em dólares é {valor_ganho_dolares}")

# 11. Peça 2 valores apostados inteiros e um valor ganho real. Calcule e mostre:
valor1 = int(input("Digite o primeiro valor apostado: "))
valor2 = int(input("Digite o segundo valor apostado: "))
valor_ganho = float(input("Digite o valor ganho: "))

# o produto do dobro do primeiro valor apostado com metade do segundo.
produto = (2 * valor1) * (valor2 / 2)
print(f"O produto do dobro do primeiro valor apostado com metade do segundo é {produto}")

# a soma do triplo do primeiro valor apostado com o valor ganho.
soma = (3 * valor1) + valor_ganho
print(f"A soma do triplo do primeiro valor apostado com o valor ganho é {soma}")

# o valor ganho elevado ao cubo.
valor_ganho_cubo = valor_ganho ** 3
print(f"O valor ganho elevado ao cubo é {valor_ganho_cubo}")

# 12. Calcule o possível retorno com base na fórmula: valor_apostado * odd
valor_apostado = float(input("Digite o valor apostado: "))
odd = float(input("Digite a odd: "))
retorno = valor_apostado * odd
print(f"O possível retorno é {retorno}")

```

# **Desafio**

- Faça um Programa que pergunte quanto você ganhou em apostas esportivas online no mês e o valor total apostado. Calcule e mostre o lucro líquido no referido mês, sabendo-se que são descontados 10% para a plataforma de apostas, 5% para taxas de transação e 2% para impostos, faça um programa que nos dê:

    - - lucro bruto.
    - - quanto pagou para a plataforma de apostas.
    - - quanto pagou em taxas de transação.
    - - quanto pagou em impostos.
    - - o lucro líquido.
    - - calcule os descontos e o lucro líquido, conforme a tabela abaixo:

    + Lucro Bruto : R$
    - Plataforma de Apostas (10%) : R$
    - Taxas de Transação (5%) : R$
    - Impostos (2%) : R$
    = Lucro Líquido : R$

    Obs.: Lucro Bruto - Descontos = Lucro Líquido.

## Resolução

```python

ganho_mensal = float(input("Digite quanto você ganhou em apostas esportivas online no mês: "))
valor_total_apostado = float(input("Digite o valor total apostado no mês: "))

# Cálculos
lucro_bruto = ganho_mensal - valor_total_apostado
plataforma_apostas = lucro_bruto * 0.10
taxas_transacao = lucro_bruto * 0.05
impostos = lucro_bruto * 0.02
descontos = plataforma_apostas + taxas_transacao + impostos
lucro_liquido = lucro_bruto - descontos

# Resultados
print(f"+ Lucro Bruto : R$ {lucro_bruto:.2f}")
print(f"- Plataforma de Apostas (10%) : R$ {plataforma_apostas:.2f}")
print(f"- Taxas de Transação (5%) : R$ {taxas_transacao:.2f}")
print(f"- Impostos (2%) : R$ {impostos:.2f}")
print(f"= Lucro Líquido : R$ {lucro_liquido:.2f}")
```
